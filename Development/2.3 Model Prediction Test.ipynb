# Model Prediction Test

```python
import warnings
warnings.filterwarnings(action='ignore')
from keras.models import load_model
import numpy as np
import pandas as pd
import itertools
```

Load test dataset and predict test

```python
model = load_model('./model/model_weights.h5')
test_x = np.load('./word_index_list_test.npy')
test_y = model.predict(test_x)
```

Load true labels

```python
labels = np.load('./labels_test.npy')
```

Summarize the result predictions in a data frame

```python
result = pd.DataFrame(test_y)
result.columns = ['equity', 'fixed_income', 'derivatives', 'alternatives']
result['predicted_ label'] = result.apply(lambda x: np.argmax(x[ :-1]), axis=1)
result['label'] = labels
```

save prediction results

```python
result.to_csv('./results.csv')
```
